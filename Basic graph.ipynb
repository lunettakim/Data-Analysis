```python
# importing Bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

# prepare some data
X = [1,2,3,4,5]
Y = [6,7,8,9,10] # lists should have same length for bokeh

# prepare the output file
output_file("Line.html")

# create a figure object
f = figure()

# create line plot
f.line(X,Y)

# write the plot in the figure object
show(f)
```

```python
x = [1,3,5,7,9,11]
y = [2,7,9,11,13,15]

output_file("new.html")
new = figure()
new.circle(x,y)
show(new)
```

```python
# Use Bokeh with pandas

# importing Bokeh and pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

# prepare some data
df = pandas.read_csv("data.csv")
X = df["x"]
Y = df["y"]

# prepare the output file
output_file("line_fromcsv.html")

# create a figure object
f = figure()

# create line plot
f.line(X,Y)

# write the plot in the figure object
show(f)
```

```python
# Add style of the plot
import pandas
from bokeh.plotting import figure, output_file, show

df = pandas.read_csv("bachelors.csv")
X = df["Year"]
Y = df["Social Sciences and History"]

output_file("bachelors.html")
p = figure(plot_width=500, plot_height=400, tools='pan')

p.title.text = "Bachelor degree"
p.title.text_color = "Blue"
p.title.text_font="times"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Date"
p.yaxis.axis_label="Intensity"  

p.line(X,Y)

show(p)
```

```python


import pandas

from bokeh.plotting import figure, output_file, show

df=pandas.read_excel("verlegenhuken.xlsx")
df["Temperature"] = df["Temperature"]/10
df["Year"] = df["Year"]/10

output_file("temp.html")

f = figure(plot_width=700, plot_height=500, tools='pan')

f.title.text = "Temerature and Air pressure"
f.title.text_font="arial"
f.title.text_color="Red"
f.title.text_font_style="italic"

f.xaxis.axis_label = "Temperature"
f.yaxis.axis_label = "Pressure"

f.circle(df["Temperature"],df["Pressure"],size=0.5)

show(f)
```

```python
# creating scatter plot with more visual attributes

from bokeh.plotting import figure, output_file, show

output_file("Scatter_plotting.html")
p = figure(plot_width = 500, plot_height=400, tools = 'pan, reset')
p.title.text = "Earthquakes"
p.title.text_color = "Orange"
p.title.text_font = "times"
p.title.text_font_style = "italic"

p.yaxis.minor_tick_line_color = "Yellow"

p.xaxis.axis_label = "Times"
p.yaxis.axis_label = "Value"

p.line([1,2,3,4,5],[5,6,5,5,3],line_width=2,color="Yellow",alpha=0.5 )
p.circle([i*2 for i in [1,2,3,4,5]],[5,6,5,5,3], size = [i*2 for i in [8,12,14,15,20]],
        color = "red", alpha=0.5)

show(p)
```

```python
# Time-series plot

from bokeh.plotting import figure, output_file, show
import pandas

df = pandas.read_csv("adbe.csv", parse_dates=["Date"])

p = figure(width=500, height=250, x_axis_type="datetime",
          sizing_mode="scale_both")

p.line(df["Date"], df["Close"], color="Orange", alpha=0.5)

output_file("Timeseries.html")
show(p)
```
