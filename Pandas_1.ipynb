```python
# You want to convert an entire column of a data fram into latitude, longtitude
df = pd.read_csv("supermarkets.csv")
df
```

```python
df["Address"] = df["Address"]+", "+df["City"]+", "+df["State"]+", "+df["Country"]
df
```

```python
df["Coordinates"]=df["Address"].apply(nom.geocode) #pandas method "apply"
df
```

```python
df.Coordinates[0]
```

```python
df.Coordinates[0].latitude
```

```python
df["Latitude"]=df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df["Longitude"]=df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
# storing all value in temporary x variable and then for each valuable, we apply
# latitude.
# if ~ means, apply latitude if x is not None (in case there is None cell)
df
```
